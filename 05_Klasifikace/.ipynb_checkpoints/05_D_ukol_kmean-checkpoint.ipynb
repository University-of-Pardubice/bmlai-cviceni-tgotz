```python
X = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

# Zjištění parametru modelu k-mean
- vykreslení elbow grafu
- zjištění ideálního počtu clusterů

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow metoda')
plt.xlabel('Počet clusterů')
plt.ylabel('WCSS (součet čtverců vzdáleností)')
plt.grid(True)
plt.show()
```

Optimální počet clusterů bude 5

# k-mean model pro zjištěný počet clusterů
- natrénování modelu
- predikce dat

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)

kmeans.fit(X)

clusters = kmeans.predict(X)

X['Cluster'] = clusters
```

# Vizualizace modelu
- vykreslení obarvených skupin grafů

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.scatter(X[X['Cluster'] == 0]['Annual Income (k$)'], X[X['Cluster'] == 0]['Spending Score (1-100)'], label='Cluster 0')
plt.scatter(X[X['Cluster'] == 1]['Annual Income (k$)'], X[X['Cluster'] == 1]['Spending Score (1-100)'], label='Cluster 1')
plt.scatter(X[X['Cluster'] == 2]['Annual Income (k$)'], X[X['Cluster'] == 2]['Spending Score (1-100)'], label='Cluster 2')
plt.scatter(X[X['Cluster'] == 3]['Annual Income (k$)'], X[X['Cluster'] == 3]['Spending Score (1-100)'], label='Cluster 3')
plt.scatter(X[X['Cluster'] == 4]['Annual Income (k$)'], X[X['Cluster'] == 4]['Spending Score (1-100)'], label='Cluster 4')

plt.scatter(kmeans.cluster_centers_[:,1], kmeans.cluster_centers_[:,2], s=200, c='black', label='Centroidy', marker='X')

plt.title('Segmentace zákazníků podle příjmu a skóre utrácení')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.scatter(X[X['Cluster'] == 0]['Age'], X[X['Cluster'] == 0]['Spending Score (1-100)'], label='Cluster 0')
plt.scatter(X[X['Cluster'] == 1]['Age'], X[X['Cluster'] == 1]['Spending Score (1-100)'], label='Cluster 1')
plt.scatter(X[X['Cluster'] == 2]['Age'], X[X['Cluster'] == 2]['Spending Score (1-100)'], label='Cluster 2')
plt.scatter(X[X['Cluster'] == 3]['Age'], X[X['Cluster'] == 3]['Spending Score (1-100)'], label='Cluster 3')
plt.scatter(X[X['Cluster'] == 4]['Age'], X[X['Cluster'] == 4]['Spending Score (1-100)'], label='Cluster 4')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,2], s=200, c='black', label='Centroidy', marker='X')

plt.title('Segmentace zákazníků podle věku a skóre utrácení')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.grid(True)
plt.show()
```

## Výklad modelu

- Na jaké zákazníky se zaměřit?

No podle dostupných dat bych se zaměřil na lidi od 20 do 40 let, protože ty očividně utrácejí nejvíce. Tzn cluster 2 4 a 1
